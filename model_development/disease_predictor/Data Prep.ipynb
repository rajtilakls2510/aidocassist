```python
import numpy as np
import pandas as pd
```

```python
severity = pd.read_csv(DISEASE_PRED_PATH+"/Symptom-severity.csv")
dataset = pd.read_csv(DISEASE_PRED_PATH+"/dataset.csv")
desc = pd.read_csv(DISEASE_PRED_PATH+"/symptom_Description.csv")
precaution = pd.read_csv(DISEASE_PRED_PATH+"/symptom_precaution.csv")
```

# Severity

```python
severity.head()
```

```python
severity.info()
```

```python
severity.weight.unique()
```

```python
severity.drop_duplicates(subset="Symptom", keep='first', inplace=True)
```

```python
severity.replace(1,2, inplace=True)
```

```python
severity["Type"]=np.where(severity["weight"]==2, "mild", 
                          np.where(severity["weight"]==3, "mild", 
                                   np.where(severity["weight"]==4, "moderate", 
                                            np.where(severity["weight"]==5, "moderate", 
                                                     np.where(severity["weight"]==6, "severe", 
                                                              np.where(severity["weight"]==7, "severe", np.nan))))))
```

```python
severity[severity["Type"]=="moderate"]
```

```python
severity.drop(index=132, inplace=True)
```

```python
severity.to_csv(DISEASE_PRED_PATH+"/New_symptom_severity.csv", index=False)
```

# Dataset

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset.iloc[:,1:]=dataset.iloc[:,1:].replace(" ","", regex=True)
```

```python
dataset.head()
```

```python
diseases = dataset["Disease"].values
```

```python
symptoms_diseases = dataset.iloc[:,1:].apply(lambda row: row.to_list(), axis=1).values
```

```python
training_dataset = pd.DataFrame(columns=severity['Symptom'].values, index=dataset.index)
training_dataset.replace(np.nan, 0, inplace=True)
training_dataset["disease"]=diseases
training_dataset['symptoms']=symptoms_diseases
training_dataset.head()
```

```python
# Takes time to convert all columns
for i in severity['Symptom'].values:
  training_dataset[i] = training_dataset.apply(lambda x:1 if i in x.symptoms else 0, axis=1)
```

```python
training_dataset.drop(columns="symptoms", inplace=True)
```

```python
training_dataset.drop_duplicates(inplace=True)
```

```python
training_dataset.shape
```

```python
training_dataset.to_csv(DISEASE_PRED_PATH+"/Prepared_Data2.csv", index=False)
```
